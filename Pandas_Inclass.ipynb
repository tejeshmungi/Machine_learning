## Data Normalization

Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include scaling the variable so the variable average is 0, scaling the variable so the variance is 1, or scaling variable so the variable values range from 0 to 1

```python
data[['length','height','width']]
```

There are serveral ways to Normalize Data:
    <br>1.Simple Feauture scaling xnew = xold/xmax
    <br>2.min-max xnew = xold-xmin/xmax-xmin
    <br>3.Z-Score xnew = xold-mean/std

```python
data['length'] = data['length']/data['length'].max()
data['width'] = data['width']/data['width'].max()
data['height'] = data['height']/data['height'].max()
```

```python
data[['length','height','width']]
```

## Turning Categorical varible into Numerical Variables

We see the column "fuel-type" has two unique values, "gas" or "diesel". 
Regression doesn't understand words, only numbers. 
To use this attribute in regression analysis, we convert "fuel-type" into indicator variables.

```python
data["fuel-type"].unique()
```

```python
dummy_var_1 = pd.get_dummies(data["fuel-type"])
dummy_var_1.head()
```

```python
# merge data frame "df" and "dummy_variable_1" 
data = pd.concat([data, dummy_var_1], axis=1)

# drop original column "fuel-type" from "df"
data.drop("fuel-type", axis = 1, inplace=True)
```

<h2 id="basic_grouping">Basics of Grouping</h2>

```python
data['drive-wheels'].unique()
```

#### we want to know, on average, which type of drive wheel is most valuable?

```python
df_group_one = data[['drive-wheels','body-style','price']]
df_group_one
```

```python
# grouping results
df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean()
df_group_one
```

```python
# multiple variables/labels grouping
df_gptest = data[['drive-wheels','body-style','price']]
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

```python
## Converting into Pivot table
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

```python
grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
```
